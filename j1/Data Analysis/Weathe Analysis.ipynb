# 6. Seasonal Variations
Question: How do weather patterns change by season?

````python
df.sample(2)
````

````python
df['Month'] = df['Date/Time'].dt.month
df['Season'] = df['Month'].apply(lambda x: 'Winter' if x in [12,1,2] else 
                                'Spring' if x in [3,4,5] else
                                'Summer' if x in [6,7,8] else 'Fall')

seasonal_stats = df.groupby('Season').agg({
    'Temp_C': ['mean', 'std'],
    'Rel Hum_%': 'mean',
    'Wind Speed_km/h': 'mean'
})
````

````python
df.columns
````

````python
# Step 1: Group by Season and calculate mean and std
seasonal_stats = df.groupby('Season')['Temp_C'].agg(Temp_C_mean='mean', Temp_C_std='std').reset_index()

# Step 2: Plot with error bars
plt.errorbar(seasonal_stats['Season'], 
             seasonal_stats['Temp_C_mean'], 
             yerr=seasonal_stats['Temp_C_std'], fmt='o-', capsize=5, color='orange')

plt.title('Seasonal Temperature with Standard Deviation')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
````

# Analysis of Line Chart: Extreme Weather Events by Hour of Day

This notebook analyzes a line chart that illustrates the frequency of two different types of extreme weather events—"Wind" and "Low Visibility"—distributed across the 24 hours of a day.

**The visualization shows:**
* **Title:** Extreme Weather Events by Hour of Day
* **X-axis:** The Hour of the day, from 0 (midnight) to 23 (11 PM).
* **Y-axis:** The "Number of Events," representing the total count of extreme events recorded for each hour.
* **Lines:** Two distinct lines are plotted: one for "Low Visibility" events (blue) and one for "Wind" events (orange).
* **Legend:** Clearly identifies which color corresponds to which type of extreme event.

## Key Insights from the Plot

This chart is highly effective at revealing the distinct daily (diurnal) patterns of the two different weather phenomena. The insights are best understood by analyzing each line separately and then comparing them.

### 1. Pattern of Extreme "Wind" Events (Orange Line)

* **Daytime Phenomenon:** Extreme wind events occur almost exclusively during daylight hours, roughly from 7 AM to 10 PM.
* **Bimodal Distribution:** The pattern shows two main peaks.
    * A major **morning peak around Hour 9 (9 AM)**, which is the time with the most extreme wind events overall (~31 events).
    * A second, slightly lower **afternoon peak around Hour 13 (1 PM)**.
* **Meteorological Context:** This pattern is driven by solar heating. As the sun warms the ground, the air near the surface becomes buoyant and mixes with the air higher up. This vertical mixing brings faster-moving winds from the upper atmosphere down to the surface, causing gusts and higher sustained wind speeds. This effect is strongest from mid-morning through the afternoon. The atmosphere is more stable at night, leading to calmer surface winds.
---

### 2. Pattern of "Low Visibility" Events (Blue Line)

* **Nighttime/Early Morning Phenomenon:** The pattern for low visibility is the complete opposite of the wind pattern. These events are concentrated in the overnight and early morning hours.
* **Single Peak:** The frequency peaks around **Hour 6 (6 AM)**, just around sunrise for many parts of the year.
* **Daytime Dissipation:** The number of events drops sharply after sunrise and remains at or near zero for the entire afternoon and evening.
* **Meteorological Context:** This pattern is a classic signature of **radiation fog**. At night, the ground loses heat and cools down. This, in turn, cools the layer of air directly above it to its dew point, causing water vapor to condense into fog. When the sun rises, its energy heats the ground and burns off the fog, causing visibility to improve.
---

### 3. Comparing the Two Patterns

* The two types of events are almost **inversely correlated**. The conditions that favor one inhibit the other.
* **High wind risk** occurs during the day when the atmosphere is well-mixed and turbulent.
* **High low-visibility risk** occurs at night when the atmosphere is calm, cool, and stable.
* This provides a powerful and actionable insight: the primary weather risks for this location have very distinct and predictable time windows.
